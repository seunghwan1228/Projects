```
import re
```

```
def remove_punt(x):
  pattern = re.compile('[^a-zA-Z0-9가-힣]+')
  x = re.sub(pattern, ' ', x)
  return x
```

```
remove_punt(df[0][0])
```

```
df[0] = df[0].apply(remove_punt)
```

```
from soynlp.word import WordExtractor
```

```
word_ext = WordExtractor()
```

```
courpus = df[0].to_list()
```

```
word_ext.train(courpus)
```

```
word_ext.word_scores()
```

```
import math
from soynlp.tokenizer import LTokenizer
```

```
score = word_ext.word_scores()
```

```
scores = {key:(score[key].cohesion_forward * math.exp(score[key].right_branching_entropy)) for key in score.keys()}
```

```
scores
```

```
ltoken = LTokenizer(scores=scores)
```

```
df[0][0]
```

```
ltoken.tokenize(df[0][0])
```

```
tokenized_list = []
for i in df[0]:
  tokenized_list.append(ltoken.tokenize(i))
```

```
tokenized_list[:5]
```

```
from keras.preprocessing import text, sequence
```

```
ktok = text.Tokenizer(oov_token='UNK')
```

```
ktok.fit_on_texts(tokenized_list)
```

```
len(ktok.index_word)
```

```
gc.collect()
```

```
ktok_result = ktok.texts_to_sequences(tokenized_list)
```

```
import numpy as np
length_check = []
for i in ktok_result:
  length_check.append(len(i))

print(np.max(np.array(length_check)))
```

```
del length_check
del tokenized_list
```

```
gc.collect()
```

```
ktok_pad = sequence.pad_sequences(ktok_result, maxlen=2000, padding='post')
```

```
print(ktok_pad.shape)
ktok_pad
```

```
train = ktok_pad[:58700]
test = ktok_pad[58700:]
```

```
train_y = np.ones(shape = train.shape[0])
test_y = np.ones(shape=test.shape[0])
```

```
del ktok_pad
```

```
gc.collect()
```

```
from keras import models, layers
from keras import backend as K


def man_dist_lambda(left, right):
  return K.exp(-K.sum(K.abs(left-right), axis=1, keepdims=True))
```

```
max_len = len(ktok.index_word)
```

```
emb = layers.Embedding(max_len+1, 100, input_length=2000)
l_lstm = layers.Bidirectional(layers.CuDNNLSTM(128))

q1_input = layers.Input(shape=(2000,))
q2_input = layers.Input(shape=(2000,))

q1 = emb(q1_input)
q2 = emb(q2_input)

l1 = l_lstm(q1)
l2 = l_lstm(q2)

man_dist = layers.Lambda(lambda x: man_dist_lambda(x[0], x[1]))([l1,l2])

malstm = models.Model([q1_input, q2_input], man_dist)
```

```
from keras import optimizers
optimizer = optimizers.Adadelta(clipnorm=1.25)
```

```
malstm.compile(loss='mse', optimizer=optimizer, metrics=['acc', 'mae'])
```

```
from tensorboardcolab import * 
```

```
tbc = TensorBoardColab()
```

```
from keras import backend as K
```

```
model_1 = malstm.fit([train, train], train_y, epochs=50, batch_size=128, validation_data=([test, test], test_y), callbacks=[TensorBoardColabCallback(tbc)])
```
