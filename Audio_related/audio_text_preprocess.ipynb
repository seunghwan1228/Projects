```
# Text Preprocess Function
def remove_next_space(x):
  next_line_pattern = re.compile(r'\\\\n')
  space_pattern = re.compile(r'[" "]+')
  return re.sub(space_pattern, ' ', re.sub(next_line_pattern, '', x)).strip().lstrip()

def remove_special_char(x):
  special_char_pattern = re.compile(r"""[“”……<>\[\]'‘’]+""")
  return re.sub(special_char_pattern, '', x)

# Filter Function
def remove_dash(x):
  if '--' in x:
    return None
  else:
    return x

def remove_english_num_included(x):
  eng_num_pattern = re.compile(r'[a-zA-Z0-9]+')
  if re.search(eng_num_pattern, str(x)) is not None:
    return None
  else:
    return x

def select_short_audio(x, length):
  audio_length = x.split('_')[-1]
  audio_length = float(re.sub(r'.wav', '', audio_length))
  if audio_length > length:
    return None
  else:
    return x

def get_exception(x):
  if re.search(r'-', str(x)) is not None:
    return None
  else:
    return x

def convert_audio_path(x):
  split_list = x.split('_')
  result = 'data/' + split_list[0] + '/' + str(x)
  return result
```

```
original_dir = r'C:\Users\seung\OneDrive\Desktop\AITHE_PROJECT\data'
os.chdir(original_dir)

result = []
for dir_ in os.listdir():
  # dir_ == 000, 001, 002
  dir_num = dir_
  text_file_name = dir_num + '.txt'
  dir_full_path = os.path.join(original_dir, dir_num)
  os.chdir(dir_full_path)


  text_file_name = '{}.txt'.format(dir_num)

  # Text Part
  with open(original_dir + r'\{}\{}'.format(dir_num, text_file_name), 'r', encoding='utf-8') as f:
    lines = f.readlines()

  text_lines = [line[1:] for line in lines]
  number = [num[0] for num in lines]

  df = pd.DataFrame({'number':number, 'text':text_lines})
  df['text'] = df['text'].apply(remove_next_space)
  df['text'] = df['text'].apply(remove_special_char)

  # Convert -- includes as None
  df['text'] = df['text'].apply(remove_dash)
  df['text'] = df['text'].apply(remove_english_num_included)
  df['text'] = df['text'].apply(get_exception)

  # Audio File
  wav_list = os.listdir()
  wav_list.remove(f'{text_file_name}')
  wav_list.sort()
  df['wav_file'] = wav_list
  df['wav_file'] = df['wav_file'].apply(select_short_audio, length=10)
  df['wav_file'] = df['wav_file'].apply(convert_audio_path)
  # drop na 
  df.dropna(inplace=True)

  result.append(df) 
  os.chdir(original_dir)

final_df = pd.concat(result)
final_df.reset_index(drop=True, inplace=True)
```

```
def time_extractor(x):
  audio_length = x.split('_')[-1]
  audio_length = float(re.sub(r'.wav', '', audio_length))
  return audio_length
```
