```
pad_text = sequence.pad_sequences(token_result, maxlen=100, padding='post', dtype='int16')
```

## train test split 

```
from sklearn.model_selection import train_test_split
```

```
target = df['label'].values
```

```
x_train, x_test, y_train, y_test = train_test_split(pad_text, target, test_size=0.2, random_state=42, stratify=target)
```

```
sns.countplot(y_test)
```

# Define model

```
from keras import models, layers
```

```
input_tensor = layers.Input(shape = (100,))
x = layers.Embedding(max_len+1, 200)(input_tensor)
x = layers.SpatialDropout1D(0.3)(x)
x = layers.Bidirectional(layers.LSTM(128, return_sequences=True))(x)
x = layers.Conv1D(128, 7, activation='relu', padding='same')(x)
x_1 = layers.GlobalAveragePooling1D()(x)
x_2 = layers.GlobalMaxPooling1D()(x)
x = layers.Concatenate(axis=-1)([x_1, x_2])
x = layers.Dense(128, activation='relu')(x)
x = layers.Dense(1, activation='sigmoid')(x)

model = models.Model(input_tensor, x)
```

```
model.summary()
```

```
!pip install keras-rectified-adam
```

```
from keras_radam import RAdam
from keras import optimizers
```

```
my_optimizer = RAdam(learning_rate=0.001, decay = 0.2)
```

```
model.compile(loss='binary_crossentropy', optimizer=my_optimizer, metrics=['acc'])
```

```
history = model.fit(x_train, y_train, 
                    batch_size=512, 
                    epochs=100, 
                    validation_data=(x_test, y_test),
                    )
```

```
input_tensor = layers.Input(shape = (100,))
x = layers.Embedding(max_len+1, 200)(input_tensor)
x = layers.SpatialDropout1D(0.4)(x)
x = layers.Bidirectional(layers.LSTM(128, return_sequences=True, dropout=0.3, recurrent_dropout=0.3))(x)
x = layers.Conv1D(128, 7, activation='relu', padding='valid')(x)
x = layers.Dropout(0.3)(x)
x_1 = layers.GlobalAveragePooling1D()(x)
x_2 = layers.GlobalMaxPooling1D()(x)
x = layers.Concatenate(axis=-1)([x_1, x_2])
x = layers.Dense(128, activation='relu')(x)
x = layers.Dropout(0.3)(x)
x = layers.Dense(1, activation='sigmoid')(x)

model = models.Model(input_tensor, x)
```

```
my_optimizer = optimizers.SGD(0.01, momentum=0.8, nesterov=True)
```

```
model.compile(loss='binary_crossentropy', optimizer=my_optimizer, metrics=['acc'])
```

```
history = model.fit(x_train, y_train, 
                    batch_size=512, 
                    epochs=100, 
                    validation_data=(x_test, y_test),
                    )
```
