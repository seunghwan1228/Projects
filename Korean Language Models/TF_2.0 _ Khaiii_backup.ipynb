```
def khaii_morphing(x):
  obs_sent_full = []
  for i in khm.analyze(x):
    for k in i.morphs:
      obs_sent_full.append(str(k))
  
  return obs_sent_full
```

```
# define model

input_tensor = keras.layers.Input(shape = (1000,))
x = keras.layers.Embedding(max_len+1, 200)(input_tensor)
x = keras.layers.SpatialDropout1D(0.3)(x)
x = keras.layers.Bidirectional(keras.layers.LSTM(128, return_sequences=True))(x)
x = keras.layers.Conv1D(128, 7, activation='relu', padding='same')(x)
x_1 = keras.layers.GlobalAveragePooling1D()(x)
x_2 = keras.layers.GlobalMaxPooling1D()(x)
x = keras.layers.Concatenate(axis=-1)([x_1, x_2])
x = keras.layers.Dense(128, activation='relu')(x)
x = keras.layers.Dense(17, activation='softmax')(x)

model = keras.models.Model(input_tensor, x)
```

```
model.compile(loss='categorical_crossentropy', optimizer=tf.keras.optimizers.Adam(1e-4), metrics=['accuracy'])
```

```
my_call_rlr = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.3, verbose=1)
```

```
history = model.fit(x_train, y_train, batch_size=128, epochs=100, callbacks=[my_call_rlr], validation_data = (x_test, y_test))
```
