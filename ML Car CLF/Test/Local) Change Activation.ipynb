```
from keras.applications import VGG16
from keras import models, layers
```

```
model = VGG16(input_shape = (224,224,3))
```

```
model.summary()
```

```
model.get_layer('block1_conv1')
```

```
model.get_layer('block1_conv1').activation = 'tanh'
```

```
model.get_layer('block1_conv1').activation
```

```
model.layers
```

```
model.layers[-1]
```

```
empty_index_conv2d = []

for index, i in enumerate(model.layers):
  if 'Conv2D' in str(i):
    print(index, i)
    empty_index_conv2d.append(index)
    
print(empty_index_conv2d)
```

```
empty_index_conv2d
```

```
model.get_layer(index=1)
```

```
for i in empty_index_conv2d:
  print(i)
  print(model.get_layer(index=i).activation)
```

```
model.get_layer()
```

```
simple_model = models.Sequential()
simple_model.add(layers.Dense(10, activation = 'relu', input_shape=(100,)))
simple_model.add(layers.Dense(20, activation = 'relu'))
```

```
simple_model.summary()
```

```
from keras import activations
```

```
simple_model.get_layer(index=0).activation
```

```
simple_model.get_layer(index=0).activation = activations.tanh
```

```
simple_model.get_layer(index=0).activation
```

```
simple_model.get_layer(index=1).activation
```

```
simple_model.get_layer(index=1).activation = activations.softmax
```

```
simple_model.get_layer(index=1).activation
```

```
simple_model.compile(loss = 'categorical_crossentropy', optimizer = 'adam')
```

```
simple_model.summary()
```

```
simple_model.layers
```

```
simple_model.get_layer(index=0).activation
```
