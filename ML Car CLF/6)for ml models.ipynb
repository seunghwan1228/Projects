```
#image to 1d array with (224, 224,3) size

all_image_array = {}

for one_image in img_path:
  img_test = image.load_img('/content/train/train_data/'+f'{one_image}', target_size=(224,224,3))
  img_array = image.img_to_array(img_test)
  img_array_1d = img_array.reshape(224*224*3,)
  all_image_array[one_image] = img_array_1d
```
