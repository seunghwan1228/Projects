```
import tensorflow as tf
import numpy as np

print(tf.__version__)
```

```
def cn(shape):
  return tf.random.uniform(shape=shape)
```

Dot

```
cn((1,2,3))
```

```
cn(())
```

```
cn((1,))
```

```
h_t = cn((2, 5))
```

```
h_s = cn((2, 10, 5))
```

```
h_t
```

```
tf.expand_dims(h_t, 1)
```

```
tf.keras.layers.Dot(axes=[2,2])([h_s, tf.expand_dims(h_t, 1)])
```

```
# tf.keras.layers.Dot()
```

```
lambda_layer = tf.keras.layers.Lambda(lambda x: x / tf.sqrt(5.))

a = tf.keras.layers.Dot(axes=[2,2])([h_s, tf.expand_dims(h_t, 1)])
```

```
lambda_layer(a)
```

```
lambda_layer(a).numpy()[0] * np.sqrt(5.)
```

General


```
W = tf.keras.layers.Dense(10, use_bias=False)
h_t = cn((2, 10))
h_s = cn((2, 2, 10))

h_t = tf.expand_dims(h_t, 1)

print(h_t)
print('--')
print(h_s)
```

```
weighted_hs = W(h_s)

weighted_hs
```

```
h_t.shape
```

```
tf.keras.layers.Dot(axes=[2,2])([weighted_hs, h_t])
```

```
weighted_hs.numpy()
```

```
h_t.numpy().shape
```

```
tf.matmul(weighted_hs, h_t, transpose_b=True)
```
