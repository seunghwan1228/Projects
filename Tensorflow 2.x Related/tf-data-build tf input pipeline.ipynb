using with python generator

````
def count(stop):
  i = 0
  while i < stop:
    yield i
    i += 1
````

````
x = count(10)
````

````
for i in x:
  print(i)
````

````
ds_count = tf.data.Dataset.from_generator(count, args=[25], output_types=tf.int32, output_shapes=())
````

````
ds_count
````

````
for i in ds_count:
  print(i)
````

````
ds_count = ds_count.repeat().batch(10).take(10)
````
