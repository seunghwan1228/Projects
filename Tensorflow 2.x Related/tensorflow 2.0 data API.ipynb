```
@tf.function  # tensorflow decorator
def preprocess(line):
  X_mean = df.describe().loc['mean', :].values
  X_std = df.describe().loc['std', :].values
  n_inputs= 9
  # defs = [[0.] * n_inputs + [tf.constant([], dtype=tf.float32), ['a'] + [tf.constant([], dtype=tf.string)]]]
  defs = [tf.constant([], dtype=tf.float32),
         tf.constant([], dtype=tf.float32),
         tf.constant([], dtype=tf.float32),
         tf.constant([], dtype=tf.float32),
         tf.constant([], dtype=tf.float32),
         tf.constant([], dtype=tf.float32),
         tf.constant([], dtype=tf.float32),
         tf.constant([], dtype=tf.float32),
         tf.constant([], dtype=tf.float32),
         tf.constant([], dtype=tf.string)]

  fields = tf.io.decode_csv(line, defs)
  # return fields
  x = tf.stack(fields[:-2])
  x_cat = tf.stack(fields[-1:])
  y = tf.stack(fields[-2:-1])

  return (x - X_mean) / X_std, x_cat, y
```

```
counter = 0
for i in dataset:
  print(i)

  counter += 1
  if counter == 2:
    break
```

```
dataset = dataset.map(preprocess)
```
