```
import numpy as np
import tensorflow as tf
```

```
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.imdb.load_data(num_words=10000)
```

```
x_train = tf.keras.preprocessing.sequence.pad_sequences(x_train, maxlen=150, padding='post')
x_test = tf.keras.preprocessing.sequence.pad_sequences(x_test, maxlen=150, padding='post')
```

```
class Attention(tf.keras.layers.Layer):
  def __init__(self, units):
    super(Attention, self).__init__()
    self.units = units

    self.w = tf.keras.layers.Dense(units)
    self.u = tf.keras.layers.Dense(1, use_bias=False)

  def call(self, inputs):
    u_i = self.w(inputs)
    u_i = tf.nn.tanh(u_i)
    a_i = self.u(u_i)
    a_i_nn = tf.nn.softmax(a_i)
    v = inputs * a_i_nn
    v = tf.reduce_sum(v, axis=1)

    return v
```

```
att = Attention(256)
```

```
input_ = tf.keras.layers.Input(shape=(150,))
x = tf.keras.layers.Embedding(10000+1, 300, mask_zero=True)(input_)
x = tf.keras.layers.GRU(256, return_sequences=True)(x)
x = att(x)
x = tf.keras.layers.Dense(128, activation='relu')(x)
x = tf.keras.layers.Dense(1, activation='sigmoid')(x)
```

```
model = tf.keras.models.Model(input_, x)
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])
```

```
model.summary()
```

```
hist = model.fit(x_train, y_train, batch_size=256, epochs=50, validation_data=(x_test, y_test))
```

```
class Attention2(tf.keras.layers.Layer):
  def __init__(self, units, **kwargs):
    super(Attention2, self).__init__(**kwargs)
    self.units = units

    self.w = tf.keras.layers.Dense(units, name='w')

  def build(self, input_shape):
    self.u = self.add_weight(shape=(input_shape[-1], ),
                             initializer='random_normal',
                             trainable=True)

  def call(self, inputs):
    u_i = self.w(inputs)
    u_i = tf.nn.tanh(u_i)


    u = tf.expand_dims(self.u, axis=1)
    # a_i_k = tf.keras.backend.dot(u_i, u)
    a_i_tf = tf.matmul(u_i, u)
    a_i_nn = tf.nn.softmax(a_i_tf)
    v = inputs * a_i_nn
    v = tf.reduce_sum(v, axis=1)

    return v
```
